```python
from newspaper import Article

def extract_and_view_articles(url):
    # Initialize a newspaper article object
    article = Article(url)
    
    try:
        # Download the article
        article.download()
        
        # Parse the article
        article.parse()
        
        # Print the title and text of the article
        print("Title:", article.title)
        print("Text:", article.text)
        print("-" * 50)
    
    except Exception as e:
        print("Error occurred while extracting article:", e)

# URL of the website containing the articles
url = "https://www.cnbc.com/2024/03/06/microsoft-ai-engineer-says-copilot-designer-creates-disturbing-images.html"

# Extract and view articles from the provided URL
extract_and_view_articles(url)
```
